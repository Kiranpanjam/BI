```python
# Load libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

```python
cars = pd.read_csv("carAuction.csv")
```

```python
cars.head()
```

```python
cols = ['Auction','Color','MMRCurrentAuctionAveragePrice','Size','TopThreeAmericanName','VehBCost','VehicleAge','VehOdo','WarrantyCost','WheelType']
```

```python
X = cars[cols]
```

```python
y = cars.IsBadBuy
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
y = le.fit_transform(y)
```

```python
y
```

```python
X = X.apply(le.fit_transform)
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
print("Yes",y_train.sum() / y_train.size)
print("No ",(y_train.size-y_train.sum()) / y_train.size)
```

```python
print("Yes",y_test.sum() / y_test.size)
print("No ",(y_test.size-y_test.sum()) / y_test.size)
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier(criterion = "entropy", random_state = 100, 
            max_depth = 2, max_leaf_nodes=8)

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)

#Predict the response for train dataset
z_pred = clf.predict(X_train)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Accuracy:",metrics.accuracy_score(y_train, z_pred))
```

```python
metrics.confusion_matrix(y_test,y_pred)
```

```python
metrics.confusion_matrix(y_train,z_pred)
```

```python
from sklearn.metrics import precision_recall_fscore_support as score

predicted = y_pred
y_test = y_test

precision, recall, fscore, support = score(y_test, predicted)

print('precision: {}'.format(precision))
print('recall: {}'.format(recall))
print('fscore: {}'.format(fscore))
print('support: {}'.format(support))
```

```python
from sklearn.metrics import precision_recall_fscore_support as score

predicted = z_pred
y_test = y_train

precision, recall, fscore, support = score(y_test, predicted)

print('precision: {}'.format(precision))
print('recall: {}'.format(recall))
print('fscore: {}'.format(fscore))
print('support: {}'.format(support))
```
